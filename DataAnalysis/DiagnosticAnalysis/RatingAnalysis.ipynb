# Analyze Rating


## Question 1
    Does rating of the app affected by whether the app support ads or not? 

## Read Data

```python
# read data as dataframe from csv file
df = pd.read_csv('../../Dataset/Preprocessed_data.csv', error_bad_lines=False)
print(df.shape)
```

```python
df.head()
```

## Clean

```python
df = convert_to_numeric(df)
df=convert_binary(df)
df.head(10)
```

## Select Interesting  Variables

```python
features_df = df[[ 'Ad Supported', 'Rating', 'Content Rating','Minimum Android']]
print(features_df.shape)
```

### Add rating value column to represent real representation for the rating

```python
#convert rating count column to float
df['Rating Count'] = df['Rating Count'].astype(float)
# add a new column called Rating_Value which equal rating * rating count
df['Rating_Value'] = df['Rating'] * df['Rating Count']
```

```python
features_df.head(10)
```

## Rating values distribution
    from the histogram we found that 0 rating is max value, that tends 
    to have the meaning of no ratings not 0 rating so it' better to drop
    it as it affects the mean of rating a lot

```python
plt.hist(df['Rating'])
```

```python
df=df.assign(IsRated= df['Rating'].apply(lambda x: 1 if x *1 > 0 else 0))
```

```python
print(df.columns)
```

```python
apps_rated = df.groupby('Ad Supported')['IsRated'].sum().head(10)
# plot pie chart for the number of apps rated and not rated for ad supported apps and not ad supported apps
plt.pie(apps_rated, labels=['no Ads', 'support Ads'], autopct='%1.1f%%',startangle=90, colors=['#ff9999','#66b3ff'])
# set the title for the figure
plt.title("which apps got rated more?", fontsize=20)
# show the figure
plt.show()
```

### Mean of app ratings for ad supported/ not supported

#### 1- assuming zeros are ratings

```python
app_rating = df.groupby('Ad Supported')['Rating_Value'].mean().sort_values(ascending=False).head(10)
app_rating.nlargest(15).plot(kind='bar', figsize=(7, 4))
```

#### 2- assuming zeros are no rating

```python
df = df[df['Rating'] != 0.0]
app_rating = df.groupby('Ad Supported')['Rating_Value'].mean().sort_values(ascending=False).head(10)
app_rating.nlargest(15).plot(kind='bar', figsize=(7, 4))
```

### Results
  - unlike our expectations ratings are not much affected by ads.
  - in both cases there's no significance difference between mean of rating in the 2 groups.


## Question 2
    Is there a relation between rating and last update?

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
df['Last Updated'] = df['Last Updated'].dt.year
print("Unique values in Last Updated: ", df['Last Updated'].unique())
```

```python
df['Released'] = pd.to_datetime(df['Released'])
df['Released'] = df['Released'].dt.year
print("Unique values in Released: ", df['Released'].unique())
```

```python
content_rating = df.groupby('Last Updated')['Rating_Value'].mean().sort_values(ascending=False).head(10)
content_rating.nlargest(15).plot(kind='bar', figsize=(7, 4))
```

```python
sub_blot(df, "Released", "Last Updated", "Rating_Value", "Mean Ratings by Last Updated for each category")
```

## Results
- the year with highest rating is not the latest one so, last up date has no effect on rating.


## Question 3
    Is there a relation between rating and content rating?

```python
#get unique values of content rating
content_rating = df.groupby('Content Rating')['Rating'].mean().sort_values(ascending=False).head(10)
content_rating.nlargest(15).plot(kind='bar', figsize=(7, 4), title='Mean Rating for each content rating')
```

```python
sub_blot(df, "Category", "Content Rating", "Rating", "Mean Ratings by Content Rating for each category")
```

### Results:
- there's no correlation between rating of app and its last update.
- as expected most of tne content ratings unrated.
- for specific app category there's a content rating type that has highest
  rating than other ,so company can know which category of people prefer which 
  app category.
