```python
import pandas as pd

df_all = pd.read_csv('../../../Dataset/Preprocessed_data.csv', error_bad_lines=False)
```

```python
df_all.head()
```

```python
df_all_copy = df_all.copy()
df_all_copy.head()
```

```python

df_all_copy['Installs'] = df_all_copy['Installs'].astype('str')
df_all_copy.Installs
```

```python

df_all_copy['Installs'] = df_all_copy['Installs'].str.replace(r'[^0-9]', '0')
# df_all_copy['Installs'] = df_all_copy['Installs'].str.replace(r'..', '0')
df_all_copy.Installs
```

```python
df_all_copy.Installs = df_all_copy.Installs.astype(float)
df_all_copy.Installs.mean()
```

```python
df_cat_avg_installs = df_all_copy.groupby(['Category'], group_keys=True).apply(lambda x: x)
df_cat_avg_installs.head()
```
